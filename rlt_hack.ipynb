```python
prod.loc[prod['utm_source'].isin(['vkontakte', 'Telegram', 'etp_roseltorg']), 'Канал трафика'] = 'VK, таргетинг,р'
prod.loc[prod['utm_source'].isin(['push', 'posteruc', 'qr', 'QR', 'vebinar', ' booklet', '']), 'Канал трафика'] = 'Промо-материалы,р'
prod.loc[prod['utm_source'].isin(['gos.roseltorg.ru', 'com.roseltorg.ru', '178fz', 'flocktory', 'partner']), 'Канал трафика'] = 'Партнерский маркетинг,р'
prod.loc[prod['utm_source'].isin(['yandex.promopages', 'dzen_suggest_ad_priority', 'dzen', 'dzen_suggest_nav_priority']), 'Канал трафика'] = 'Я.Дзен,р'
prod.loc[prod['utm_source'].isin(['mindbox', 'email', 'emarsys']) | (prod['utm_medium'].isin(['email', 'mail'])), 'Канал трафика'] = 'Email-рассылка, сервис,р'
prod.loc[prod['utm_source'].isin(['sms']) | (prod['utm_medium'].isin(['CoMagic'])), 'Канал трафика'] = 'whatsapp, рассылка,р'
prod.loc[prod['utm_source'].isin(['yandex', 'yandex-search', 'yandex-network', 'direct', 'yandex-direct']) & (prod['utm_medium'].isin(['cpc', 'cpa'])), 'Канал трафика'] = 'Я.Директ,р'
prod.loc[prod['utm_medium'].isin(['organic']), 'Канал трафика'] = 'СЕО, контенкт,р'
prod.loc[prod['utm_medium'].isin(['referral']), 'Канал трафика'] = 'Реферальный трафик'
prod.loc[prod['utm_source'].isin([('(direct')]) & (prod['utm_medium'].isin(['(none)'])), 'Канал трафика'] = 'Прямой трафик'
```
